#### Traveling Salesperson Algorithm

```python
#!/usr/bin/env python3
"""Run and time a nearest neighbor TSP solution."""

import sys
import timeit
import math



RUNS = 3

data = "[['21','41'],['13','14'],['1224','12']]"

print("data-type: ", type(data),"data: ", data)

def closestpoint(point, route):
    dmin = float("inf")
    for p in route:
        d = math.sqrt((float(point[0]) - float(p[0]))**2 + (float(point[1]) - float(p[1]))**2)
        if d < dmin:
            dmin = d
            closest = p
    return closest, dmin




def getFastRoute(data):

    point, *route = data
    path = [point]
    sum = 0
    
    while len(route) >= 1:
        closest, dist = closestpoint(path[-1], route)
        path.append(closest)
        route.remove(closest)
        sum += dist

    # Go back the the beginning when done.
    closest, dist = closestpoint(path[-1], [point])
    path.append(closest)
    sum += dist

    #print("Optimal route:", path)
    #print("Length:", sum)
    return path
```
